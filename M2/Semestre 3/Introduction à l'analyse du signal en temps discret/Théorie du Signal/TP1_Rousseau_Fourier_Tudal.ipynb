```python
### Importation des Modules

import numpy as np
import matplotlib.pyplot as plt
import math as m
```

```python
### Definition des Parametres

A=1  #Amplitude
T=1   #Periode
tau=0.5  #duree du pic

### Fonction Signal

def u(t):
    return A*((t%T<(tau/2))|((t%T>(T-tau/2))))

### Application

tps=np.linspace(-T,3*T,1000)
U=u(tps)

plt.plot(tps,U)
```

```python
### Calcul des coefficients

per=np.linspace(0,T,1000)  # creation des temps sur une periode
pas=per[1]-per[0]  # pas

### Fonctions coefficients

def a(n):
    return sum(u(per)*np.cos(2*m.pi*n*per/T)*pas)*2/T*(1/2*(n==0)+1*(n!=0))
    
def b(n):
    return sum(u(per)*np.sin(2*m.pi*n*per/T)*pas)*2/T
    
    
```

```python
### Construction du signal par somme de Sin Cos

def v(t,n=1):
    def cst(t0):
        return a(0)
    vi=cst(t)
    for i in range(1,n+1):
        vi+=a(i)*np.cos(2*m.pi*i*t/T)+b(i)*np.sin(2*m.pi*i*t/T)
    return vi
        
```

```python
### Affichage

tps=np.linspace(-T,3*T,1000)
U=u(tps)
V=v(tps,20)
plt.plot(tps,U)
plt.plot(tps,V)
```
